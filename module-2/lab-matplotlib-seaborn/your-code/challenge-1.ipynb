Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
bar_chart_df = vehicles.pivot_table(index = 'Fuel Type', values = 'City MPG', aggfunc = 'mean').reset_index().sort_values(by = 'City MPG', ascending = True)
bar_chart = bar_chart_df.plot('Fuel Type', 'City MPG', kind = 'barh')

bar_chart.set_xlabel('AVG City MPG')
bar_chart.set_title('AVG City MPG by Fuel Type')
```
