```python
#Challenge 3: Hidden Cold Joke

# pulling scavenger repo api
request = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents', auth=gh_session.auth)
results = request.json()
df = pd.DataFrame(results)

folder_names = list(df['name'])
folder_names.remove('.gitignore')
folder_names


#Getting scavenger file paths
scavenger_file_path = []

for folder in folder_names:
    request = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/' + str(folder), auth=gh_session.auth)
    results = request.json()
    for dict1 in results:
        if 'scavengerhunt' in dict1['name']:
            scavenger_file_path.append(dict1['path'])
#     if 'scavengerhunt' in results[0]['path']:
#         scavenger_file_path.append(results[0]['path'])

scavenger_file_path
```

```python
#Sorting files
sorted_paths = [[i[7:11],i] for i in scavenger_file_path]

df = pd.DataFrame(sorted_paths)
df = df.sort_values(by=[0])
df
```

```python
#Getting scavenger file contents
scavenger_file_contents = []
    
for path in df[1]:
    request = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents/' + str(path), auth=gh_session.auth)
    results = request.json()
    scavenger_file_contents.append(results['content'])
    
scavenger_file_contents
```

```python
#Decoding encrypted strings

decoded_words = [base64.b64decode(i) for i in scavenger_file_contents]

#Cleaning content

clean_contents = []

for string in decoded_words:
    rem_bytes = string.decode('utf-8')
    rem_quotes = re.sub("'", '', str(rem_bytes))
    rem_r_space = rem_quotes.rstrip()
    clean_contents.append(rem_r_space)
    
#Printing cold joke
    
print(' '.join(clean_contents))
```
